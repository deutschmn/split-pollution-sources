```python
data = data_prep.load_beijing_data()
```

# Wind analysis

```python
device_id = '1037A'
start_date = '2020-01-01'
end_date = '2020-01-15'

df = data.measurements
df = df[(df['device_id'] == device_id) & (df['time'] >= start_date) & (df['time'] <= end_date)]
df.index = df['time']
df = df[['u_component_of_wind', 'v_component_of_wind']]

plt.figure(figsize=(20, 5))
plt.yticks(None)
plt.quiver(df.index, 0, df['u_component_of_wind'], df['v_component_of_wind'], width=0.001, alpha=0.5)
plt.title(f"wind @ {device_id}")
plt.yticks([],[])
plt.show()
```
